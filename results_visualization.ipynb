```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_rel

```

```python
def plot_hyperparameters(file_path):
    df = pd.read_csv(file_path)
    df.sort_values(by='mean_train_score', ascending=True, inplace=True, ignore_index=True)
    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df['mean_train_score'], label='Train Accuracy')
    plt.plot(df.index, df['mean_test_score'], label='Test Accuracy')
    plt.legend()
    plt.xlabel('Hyperparameters')
    plt.ylabel('Accuracy')
    plt.title(f'Train and Test Accuracy {file_path.split("/")[-1]}')
```

```python
#load from csv
file_path = 'results/XGBoost_SMOTETomek_results.csv'
plot_hyperparameters(file_path)
```

```python
file_path = 'results/DecisionTree_results.csv'
plot_hyperparameters(file_path)
```

```python
for file in os.listdir('results/'):
    if file.endswith('.csv'):
        plot_hyperparameters('results/' + file)
```
